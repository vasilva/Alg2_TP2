```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data.csv")
data = data.sort_values(by="Graph Size (|V|)")
data = data.reset_index(drop=True)
data
```

```python
plt.figure(figsize=(10, 6))
plt.plot(
    data["Graph Size (|V|)"],
    data["BB Time (s)"],
    label="Branch and Bound",
    color="b",
    marker="o",
    alpha=0.7,
)
plt.plot(
    data["Graph Size (|V|)"],
    data["TAT Time (s)"],
    label="Twice Around the Tree",
    color="g",
    marker="s",
    alpha=0.7,
)
plt.plot(
    data["Graph Size (|V|)"],
    data["Ch Time (s)"],
    label="Christofides",
    color="r",
    marker="^",
    alpha=0.7,
)
timelimit = 1800
plt.axhline(y=timelimit, color="y", linestyle="--", label="Tempo limite (30min)")
plt.xlabel("Tamanho do Grafo |V|")
plt.ylabel("Tempo em segundos")
plt.title("Tempo x Tamanho do Grafo")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(
    data["Graph Size (|V|)"], data["BB Time (s)"], color="b", marker="o", alpha=0.7
)
timelimit = 1800
plt.axhline(y=timelimit, color="y", linestyle="--", label="Tempo limite (30min)")
plt.xlabel("Tamanho do Grafo |V|")
plt.ylabel("Tempo em segundos")
plt.title("Tempo x Tamanho do Grafo (Branch and Bound)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(
    data["Graph Size (|V|)"], data["TAT Time (s)"], color="g", marker="s", alpha=0.7
)
plt.xlabel("Tamanho do Grafo |V|")
plt.ylabel("Tempo em segundos")
plt.title("Tempo x Tamanho do Grafo (Twice Around the Tree)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(
    data["Graph Size (|V|)"], data["Ch Time (s)"], color="r", marker="^", alpha=0.7
)
timelimit = 1800
plt.axhline(y=timelimit, color="y", linestyle="--", label="Tempo limite (30min)")
plt.xlabel("Tamanho do Grafo |V|")
plt.ylabel("Tempo em segundos")
plt.title("Tempo x Tamanho do Grafo (Christofides)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(
    data["Graph Size (|V|)"],
    data["TAT Relative"],
    label="Twice Around the Tree",
    color="g",
    marker="s",
    alpha=0.7,
)
plt.plot(
    data["Graph Size (|V|)"],
    data["Ch Relative"],
    label="Christofides",
    color="r",
    marker="^",
    alpha=0.7,
)
average_tat = data["TAT Relative"].mean()
average_ch = data["Ch Relative"].mean()
plt.axhline(
    y=average_tat,
    color="c",
    linestyle="--",
    label=f"Resultado Relativo Médio TAT (%) = {100*average_tat:.2f}%",
)
plt.axhline(
    y=average_ch,
    color="y",
    linestyle="--",
    label=f"Resultado Relativo Médio Ch (%) = {100*average_ch:.2f}%",
)

plt.xlabel("Tamanho do Grafo |V|")
plt.ylabel("Resultado Relativo")
plt.title(
    "Desempenho Relativo de Twice Around the Tree e Christofides (Valor ótimo = 100%)"
)
plt.legend()
plt.grid(True)
plt.show()
```
